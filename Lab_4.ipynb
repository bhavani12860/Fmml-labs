## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

## Questions:
1) What is the significance of using data augmentations like resize, crop etc on training data?

##ANSWER:-
Using data augmentations like resize, crop, flip, rotate, color jitter, and others during training has significant importance in improving the performance and generalization ability of machine learning models, especially in computer vision. Here’s why they matter:

🧠 1. Prevent Overfitting
Augmentations generate variations of training images, effectively creating a larger and more diverse dataset.

This helps the model not memorize specific details of training images and instead learn general features.

📈 2. Improve Generalization
Real-world data can differ slightly from training data (e.g., different lighting, angles, scales).

Augmentations simulate these variations, making the model more robust to such changes during inference.

🖼️ 3. Teach Invariance
Different augmentations help the model learn invariance to certain transformations:

Resize & Crop: Teaches the model to recognize objects at different scales and locations.

Flip/Rotate: Helps with orientation invariance.

Color Jitter: Makes the model less sensitive to lighting/contrast differences.

⚙️ 4. Simulate Real-World Scenarios
In many real applications, inputs won't be perfectly aligned or standardized.

Augmentations mimic these conditions, helping the model perform well in practice.

🧪 Example:
Without augmentations:
Model sees 10,000 images once.

With augmentations:
Model might effectively see 50,000+ unique variations of those same 10,000 images.

In summary, augmentations boost training diversity without needing more data. It’s one of the most powerful tricks to improve model performance, especially when your dataset is limited.












2) What performance do you get if you don't use pretrained resnet model (Hint: Change pretrained=False and train the model)

##ANSWER:-
When you don’t use a pretrained ResNet model (i.e., you set pretrained=False), you're starting from random weights, and that has a big impact on model performance.

Here’s what typically happens:

🔻 Performance Drop Without Pretraining
Accuracy: Significantly lower compared to pretrained models.

Training Time: Much longer to converge.

Generalization: Poorer — higher risk of overfitting or underfitting.

📊 Typical Example (on a dataset like CIFAR-10 or a small custom dataset):

Model	Accuracy (approx.)
ResNet (pretrained)	~85–95%
ResNet (no pretraining)	~50–70% (or worse)
Note: Exact numbers vary depending on dataset, training time, augmentations, optimizer settings, etc.

🧠 Why this happens:
Pretrained models (like ResNet pretrained on ImageNet) already know useful low-level features (edges, textures) and mid-level features (shapes, object parts).

When not pretrained, the model has to learn everything from scratch, which is hard, especially with small or less diverse datasets.

✅ TL;DR:
If you set pretrained=False:

Expect worse performance unless you have a large enough dataset and lots of training time.

Pretraining gives your model a head start — like learning to walk before you run.





3) If the resnet model was pre-trained on dataset significantly different than the ants vs bees data, would you still get good performance by using this pretrained model?


##ANSWER:-
🐜🐝 You're using ResNet pretrained on ImageNet
(ImageNet has 1,000 classes: cats, dogs, planes, etc.)

Now you're applying it to Ants vs Bees, a much smaller, binary classification task.
🔍 So... what if the pretraining dataset is very different?
🎯 Answer: Yes, you can still get good performance — and here's why:
✅ 1. Low-Level Features Are Universal
The early layers of CNNs like ResNet learn basic visual features:

edges, corners, color blobs, textures.

These are useful across almost all image domains, even if the categories are different.

⚙️ 2. You Fine-Tune the Rest
Even if high-level features from ImageNet aren’t directly useful for distinguishing ants vs bees, you can:

Freeze early layers (which learn general features).

Fine-tune later layers on your ants vs bees dataset.

📈 3. Empirically Proven
Transfer learning works well even when source and target tasks are not closely related — especially when:

You don’t have a lot of data.

The pretrained model has seen a massive, diverse dataset like ImageNet.

⚠️ Caveats
If the pretraining dataset is extremely different (e.g., medical images, satellite images), you might need to fine-tune more layers, or even retrain the whole network.

But even then, pretraining helps as a better initialization than random weights.

🧠 TL;DR:
Yes, you’d still get good performance using a pretrained ResNet model — even if it was trained on something very different — because general visual patterns transfer well. And you can always fine-tune to adapt it to your specific tasK.
