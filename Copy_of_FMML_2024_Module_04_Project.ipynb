### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient =df(previous_x) ## [REQUIRED]
    new_x =previous_x-step_multiplier*gradient  ## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))## [REQUIRED])

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python

import matplotlib.pyplot as plt

# Define the function whose cost we are minimizing
def f(x):
    return x**3 - 3*x  # Example function

# Gradient descent function returning x values and costs
def gradient_descent_with_cost(derivative_func, cost_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    x_history = [x]
    cost_history = [cost_func(x)]  # Track cost values
    for _ in range(max_iter):
        gradient = derivative_func(x)
        next_x = x - multiplier * gradient
        x_history.append(next_x)
        cost_history.append(cost_func(next_x))
        step_size = abs(next_x - x)
        if step_size < precision:
            break
        x = next_x
    return x_history, cost_history

# Run gradient descent with different learning rates
n = 100
low_gamma_x, low_gamma_cost = gradient_descent_with_cost(
    derivative_func=dg, cost_func=f, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n
)
mid_gamma_x, mid_gamma_cost = gradient_descent_with_cost(
    derivative_func=dg, cost_func=f, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n
)
high_gamma_x, high_gamma_cost = gradient_descent_with_cost(
    derivative_func=dg, cost_func=f, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n
)

# Plot cost reduction
plt.figure(figsize=(12, 6))
plt.plot(range(len(low_gamma_cost)), low_gamma_cost, label="Low Gamma (0.0005)", color="blue")
plt.plot(range(len(mid_gamma_cost)), mid_gamma_cost, label="Mid Gamma (0.001)", color="green")
plt.plot(range(len(high_gamma_cost)), high_gamma_cost, label="High Gamma (0.002)", color="red")
plt.title("Reduction in Cost for Each Iteration")
plt.xlabel("Iteration")
plt.ylabel("Cost (f(x))")
plt.legend()
plt.grid()
plt.show()
````

#### Mean Squared Error: a cost function for regression problems

#### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

````python
# Make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)

print('Shape of x_5 array:', x_5.shape)
print('Shape of y_5 array:', y_5.shape)
````

````python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python

import numpy as np

# Define the Mean Squared Error (MSE) function
def mse(y, predictions):
    return np.mean((y - predictions)**2)

# Define the gradient of the MSE function with respect to thetas
def grad(x, y, thetas):
    # Compute the gradients for theta_0 and theta_1
    predictions = thetas[0] + thetas[1] * x
    error = predictions - y
    grad_0 = (2 / len(x)) * np.sum(error)  # Gradient for theta_0
    grad_1 = (2 / len(x)) * np.sum(error * x)  # Gradient for theta_1
    return np.array([grad_0, grad_1])

# Example data (replace these with your actual data)
x_5 = np.array([1, 2, 3, 4, 5])  # Example x values
y_5 = np.array([2, 4, 6, 8, 10])  # Example y values (perfect linear relationship)

# Initial thetas and multiplier (learning rate)
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1] * x_5)

# Gradient descent loop
for i in range(1000):
    # Update thetas using gradient descent
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1] * x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1] * x_5))
````
